# Query-Processing

```python
def get_queried_rank(query):
    
    query_list = query.split(" ")
    query_rows = term_vectors_df[term_vectors_df['Terms'].isin(query_list)].drop(columns=['Terms'])

    # =           P(c_j)     * D_j_t1 * D_j_t2 * ... * D_j_tn   /   (Sum_i D_ji)    **    n
    weights = (1 / len(topics)) * query_rows.prod() / (term_vector_totals ** len(query_rows))
    weights /= weights.sum()
    
    ts_pagerank = (weights * pr_vectors_df.drop(columns=["ID", "Unbiased"])).sum(axis=1)
    
    return ts_pagerank
```

```python
def get_ranking(ts_pr):
    sorted_indices = ts_pr.sort_values(ascending=False).index
    ranked_df = pagelookup_df.iloc[sorted_indices]
    
    return ranked_df
```

## Query-Time Timing

```python
import time
```

### Ordinary PageRank

Assume we are just retreiving the pages in order, given by the global ranking vector.

```python
# Define the number of iterations
num_iterations = 50
execution_times_pr = np.zeros(num_iterations)

unbiased_rank_vector = pr_vectors_df["Unbiased"]

for i in range(num_iterations):
    start_time = time.time() 
    
    # Code to time
    get_ranking(unbiased_rank_vector)
    
    end_time = time.time() 
    execution_times_pr[i] = end_time - start_time  
```

```python
execution_times_pr
```

```python
pr_time_mean = execution_times_pr.mean()
pr_time_mean
# np.float64(0.6147586584091187)
```

### Topic-Sensitive PageRank
Assume we just need to take the linear combination of the rank vectors and retrieve the pages in order.

```python
# Define the number of iterations
num_iterations = 50
execution_times_tspr = np.zeros(num_iterations)

for i in range(num_iterations):
    
    # sample a random query from queries list (provided in 2002 paper)
    query = queries.sample(n=1).values[0]  
    
    start_time = time.time() 
    
    # Code to time
    get_ranking(get_queried_rank(query))
    
    end_time = time.time() 
    execution_times_tspr[i] = end_time - start_time  
```

```python
execution_times_tspr
```

```python
tspr_time_mean = execution_times_tspr.mean()
tspr_time_mean
```
